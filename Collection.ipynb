## The following is to get all the filters and pages for a specific mobile.

```python
import requests 
from bs4 import BeautifulSoup 
import csv 
   
URL = "http://phonedb.net/index.php?m=device&s=list&first=samsung"

r = requests.get(URL) 

soup = BeautifulSoup(r.content, 'html5lib') 
contentpages=[] 
#links = soup.find_all('div', attrs = {'class':'compact_list_item'})
links = soup.find_all('div', attrs = {'class':'container'})
#links = soup.find_all('b')

#for row in links[-2]:
for link in links[-2].find_all('a'):
    fullLink = link.get ('href')
    contentpages.append(fullLink)
```

```python
samsungpages = []
for i in contentpages:
    if i.find('/index.php?m=device&s=list&first=samsung&filter')!=-1 and i not in samsungpages:
        samsungpages.append(i)
```

## convering the above links to "http://phonedb.net/" format

```python
samsungwebs = []
for i in samsungpages:
    newlink = "http://phonedb.net"+i
    samsungwebs.append(newlink)
```

```python
samsungwebs
```

## Now getting mobiles from every page (links of mobiles)

```python
import requests 
from bs4 import BeautifulSoup 
import csv 
   
contentpages = []

for url in samsungwebs:
    r = requests.get(url) 

    soup = BeautifulSoup(r.content, 'html5lib') 
    
    # every link of a mobile is residing in a content_block_title, so getting that link
    links = soup.find_all('div', attrs = {'class':'content_block_title'})

    for row in links:
        for link in row.find_all('a'):
            fullLink = link.get ('href')
            contentpages.append(fullLink)
```

```python
contentpages
```

## We finally got the links of every mobile, now we'll extract the features of every mobile from its link
# This is a time consuming process.. BE PATIENT

```python
from bs4 import BeautifulSoup
import csv
import requests 

specs = []
for rii in contentpages:
    r = requests.get(rii) 
    soup = BeautifulSoup(r.content, 'html5lib')
    trs = soup.find_all('tr')
    count = 0
    ll = {}
    for tr in trs:
        tds = tr.find_all('td')
        if len(tds)==2:
            name = tds[0].get_text()
            field = tds[1].get_text().strip()
            name = name.strip()
            if name in attr:
                ll[name]=field
    specs.append(ll)
```

## Now we will convert the dictionary into a pandas dataframe

```python
speclist = list(specs)
temp = pd.DataFrame(columns=attr)
temp = temp.append(speclist, ignore_index = True, sort = False)
new = new.append(temp, ignore_index = True, sort = False)
```

## Now pushing the dataframe to Google sheets

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
```

```python
def iter_pd(df):
    for val in df.columns:
        yield val
    for row in df.to_numpy():
        for val in row:
            if pd.isna(val):
                yield ""
            else:
                yield val

def pandas_to_sheets(pandas_df, sheet, clear = False):
    if clear:
        sheet.clear()
    (row, col) = pandas_df.shape
    cells = sheet.range("A1:{}".format(gspread.utils.rowcol_to_a1(row + 1, col)))
    for cell, val in zip(cells, iter_pd(pandas_df)):
        cell.value = val
    sheet.update_cells(cells)

scope =  ['https://www.googleapis.com/auth/spreadsheets', "https://www.googleapis.com/auth/drive.file", "https://www.googleapis.com/auth/drive"]
creds = ServiceAccountCredentials.from_json_keyfile_name('client_secret.json', scope)
client = gspread.authorize(creds)
sheet = client.open("DS250_dataset").sheet1    
pandas_to_sheets(new,sheet)
```
